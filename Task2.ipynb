```python
## 1. Import Libraries and Load Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Load dataset
df = pd.read_csv('data.csv')
```

```python
## 2. Summary Statistics
summary_stats = df.describe()
print("Summary Statistics:")
print(summary_stats)
```

```python
## 3. Histograms for Numeric Features
numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
plt.figure(figsize=(15, 25))
for i, col in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols), 1, i + 1)
    sns.histplot(df[col], kde=True, color='skyblue')
    plt.title(f'Histogram of {col}')
plt.tight_layout()
plt.show()
```

```python
## 4. Boxplots for Numeric Features
plt.figure(figsize=(15, 25))
for i, col in enumerate(numeric_cols):
    plt.subplot(len(numeric_cols), 1, i + 1)
    sns.boxplot(x=df[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
plt.tight_layout()
plt.show()
```

```python
## 5. Correlation Matrix
plt.figure(figsize=(12, 8))
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
## 6. Pairplot for Selected Features
selected_cols = ['Engine HP', 'Engine Cylinders', 'highway MPG', 'city mpg', 'MSRP']
sns.pairplot(df[selected_cols], corner=True)
plt.show()
```
